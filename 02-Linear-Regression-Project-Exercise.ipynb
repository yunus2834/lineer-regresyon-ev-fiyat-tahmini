**TASK: The label we are trying to predict is the SalePrice column. Separate out the data into X features and y labels**

```python
X=df.drop("SalePrice",axis=1)
y=df["SalePrice"]
```

**TASK: Use scikit-learn to split up X and y into a training set and test set. Since we will later be using a Grid Search strategy, set your test proportion to 10%. To get the same data split as the solutions notebook, you can specify random_state = 101**

X ve y'yi bir eğitim seti ve test seti olarak ayırmak için scikit-learn'ü kullanın. Daha sonra bir Izgara Arama stratejisi kullanacağımız için test oranınızı %10 olarak ayarlayın. Çözüm not defteriyle aynı veri bölünmesini elde etmek için random_state = 101 belirtebilirsiniz.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

**TASK: The dataset features has a variety of scales and units. For optimal regression performance, scale the X features. Take carefuly note of what to use for .fit() vs what to use for .transform()**

GÖREV: Veri kümesi öznitelikleri çeşitli ölçeklere ve birimlere sahiptir. Optimum regresyon performansı için X özelliklerini ölçeklendirin. .fit() için ne kullanılacağına karşı .transform() için ne kullanılacağına dikkat edin.

```python
#class sklearn.preprocessing.StandardScaler(*, copy=True, with_mean=True, with_std=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train=scaler.transform(X_train)
scaled_X_test=scaler.transform(X_test)
```

**TASK: We will use an Elastic Net model. Create an instance of default ElasticNet model with scikit-learn**

Bir Elastic Net modeli kullanacağız. scikit-learn ile varsayılan ElasticNet modelinin bir örneğini oluşturun

```python
from sklearn.linear_model import ElasticNet
```

```python
#help(ElasticNet)
```

```python
base_elastic_model=ElasticNet(max_iter=1000000)
```

### GÖREV: Elastic Net modelinin alfa ve L1 oranı olmak üzere iki ana parametresi vardır. ElasticNet için bir sözlük parametre değerleri ızgarası oluşturun. Bu değerlerle oynamaktan çekinmeyin, unutmayın, çözüm seçenekleriyle tam olarak eşleşmeyebilirsiniz.

```python
params_grid={"alpha":[0.1,1,5,10,50,100],
             "l1_ratio":[.1,.5,7,.95,.99,1]    }
```

**TASK: Using scikit-learn create a GridSearchCV object and run a grid search for the best parameters for your model based on your scaled training data. [In case you are curious about the warnings you may recieve for certain parameter combinations](https://stackoverflow.com/questions/20681864/lasso-on-sklearn-does-not-converge)**

GÖREV: scikit-learn'ü kullanarak bir GridSearchCV nesnesi oluşturun ve ölçekli eğitim verilerinize dayalı olarak modeliniz için en iyi parametreleri bulmak üzere bir ızgara araması yapın.

```python
from sklearn.model_selection import GridSearchCV
```

```python
Grid_model=GridSearchCV(estimator=base_elastic_model,
                       param_grid=params_grid,
                       scoring="neg_mean_squared_error",
                        cv=5,verbose=1)
```

```python
Grid_model.fit(scaled_X_train,y_train)
```

**TASK: Display the best combination of parameters for your model**

```python
Grid_model.best_estimator_
```

```python
Grid_model.best_params_
```

**TASK: Evaluate your model's performance on the unseen 10% scaled test set. In the solutions notebook we achieved an MAE of $\$$14149 and a RMSE of $\$$20532**

Görünmeyen %10 ölçekli test setinde modelinizin performansını değerlendirin. Çözüm defterinde 14149 $ MAE ve 20532 $ RMSE elde ettik.

```python
Grid_model.cv_results_
```

```python
pd.DataFrame(Grid_model.cv_results_)
```

```python
y_pred=Grid_model.predict(scaled_X_test)
```

```python
mean_squared_error(y_test,y_pred)
```
